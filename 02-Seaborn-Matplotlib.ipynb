# Seaborn + Matplotlib

Normalmente não utilizamos só o matplotlib para fazer plots.  
O Seaborn já oferece algum grau de macro management que podemos aproveitar para acelerar os nossos plots.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Escolher um tipo global de plot
sns.set_theme(style="darkgrid")

# Escolher o tipo de font: DEVE SER ANTES DO CONTEXTO PARA O CONTEXTO NÃO FAZER OVEWRITE A ALGUNS SETTINGS
sns.set(font="Ubuntu")

# Escolher um contexto de apresentação
sns.set_context("talk") #notebook, paper, poster

# Dataset
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.histplot(data=df, x='flipper_length_mm')
```

No entanto, queremos ter mais controlo sobre o plot. Podemos embebê-lo num sistema de eixos de matplotlib.

```python
fig, ax = plt.subplots(figsize=(6,4))

sns.histplot(data=df, x='flipper_length_mm', ax=ax)

ax.set_title("Counts of penguin flipper lengths show\na bi-modal distribution", loc='left') # Título deve ter o takeaway do plot
ax.set_ylabel("Counts") 
ax.set_xlabel("Flipper length (mm)")
plt.subplots_adjust()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))

sns.histplot(data=df, x='flipper_length_mm', hue='species', bins=20, ax=ax)

ax.set_title("Gentoo Penguins have larger flipper lengths", loc='left', fontsize=16) # Título deve ter o takeaway do plot
ax.set_ylabel("Counts") 
ax.set_xlabel("Flipper length (mm)")

sns.move_legend(
    ax, "upper right",
    ncol=1, title=None, frameon=False,
    fontsize=12,
)

plt.subplots_adjust()
plt.show()
```
